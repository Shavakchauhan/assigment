```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
# Rading the files
df = pd.read_excel('data/comments_24_11_2022.xlsx',header=None)
df = df.astype(str)
columns_names = list(df.columns)
df['text'] = df.agg(' '.join,axis=1) # Joinng all the columns
df = df['text']
df.columns = ['text']
```

```python
def convert_nan_to_space(txt):
  """
  Converts nan in the text sentence to empty space
  """
    return txt.replace('nan',' ')
df = df.apply(convert_nan_to_space)
df= df.str.strip()
df.columns = ['text']
df = pd.DataFrame(df,columns=['text'])
df.replace('', np.nan, inplace=True)
df.dropna(inplace=True)
# Cleaning the dataset
```

```python
def decontractions(phrase):
    """decontracted takes text and convert contractions into natural form.
     ref: https://stackoverflow.com/questions/19790188/expanding-english-language-contractions-in-python/47091490#47091490"""
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"won\’t", "will not", phrase)
    phrase = re.sub(r"can\’t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)

    phrase = re.sub(r"n\’t", " not", phrase)
    phrase = re.sub(r"\’re", " are", phrase)
    phrase = re.sub(r"\’s", " is", phrase)
    phrase = re.sub(r"\’d", " would", phrase)
    phrase = re.sub(r"\’ll", " will", phrase)
    phrase = re.sub(r"\’t", " not", phrase)
    phrase = re.sub(r"\’ve", " have", phrase)
    phrase = re.sub(r"\’m", " am", phrase)

    return phrase
def preprocess(text):
    # convert all the text into lower letters
    # use this function to remove the contractions: https://gist.github.com/anandborad/d410a49a493b56dace4f814ab5325bbd
    # remove all the spacial characters: except space ' '
    text = text.replace(r'\n', ' ').replace(r'\r', '')
    text = text.lower()
    text = decontractions(text)
    text = re.sub('[^A-Za-z0-9 ]+', ' ', text)
    text = re.sub(' +',' ',text)
    return text

df['text'] = df['text'].apply(preprocess)
```

```python
df = df.drop_duplicates() # Dropping the duplicates
```

```python
df.to_csv('data/preprocessed_text.csv',index=False)
```
